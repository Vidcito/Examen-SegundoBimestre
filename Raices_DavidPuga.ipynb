```python
import numpy as np
from scipy.optimize import bisect
import matplotlib.pyplot as plt

def polynomial(x):
  """
  Evaluates the polynomial x^5 - 6x^4 + 2x^3 + 20x^2 - 27x + 10.
  """
  return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

def find_roots_bisect(polynomial, a, b, tolerance=1e-5, max_iterations=100):
    """
    Finds a root of a polynomial within the interval [a, b] using the bisection method.

    Args:
        polynomial: The polynomial function.
        a: The left endpoint of the interval.
        b: The right endpoint of the interval.
        tolerance: The tolerance for the root approximation.
        max_iterations: The maximum number of iterations.

    Returns:
        The approximate root or None if a root was not found.
    """

    if polynomial(a) * polynomial(b) >= 0:
        print("Function values at endpoints do not have opposite signs. Cannot guarantee a root exists in this range.")
        return None

    for _ in range(max_iterations):
        c = (a + b) / 2
        if polynomial(c) == 0 or (b - a) / 2 < tolerance:
            return c
        if polynomial(c) * polynomial(a) < 0:
            b = c
        else:
            a = c
    return (a+b)/2


# Plot the polynomial to visually identify approximate root locations.
x = np.linspace(-2, 6, 400)  # Adjusted interval for better visualization
y = polynomial(x)
plt.plot(x, y)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Polynomial Visualization')
plt.grid(True)
plt.show()

# Find intervals that have a sign change
intervals = [(-1, 0), (0, 1.5), (1.5, 2.5), (2.5,4), (4, 6)]

roots = []
for a, b in intervals:

    root = find_roots_bisect(polynomial, a, b)
    if root is not None:
        roots.append(root)

print("Roots found using bisection method:")
for i, root in enumerate(roots):
    print(f"Root {i+1}: {root}")
```
